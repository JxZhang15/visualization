```python
from PIL import Image # to open images
import requests # to get images from URLs
from io import BytesIO # to store images
```

```python
# Define the URL and headers
url = 'https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png'
headers = {'User-Agent': 'Mozilla/5.0'} # Pretends to be a web browser

# Get the image from the internet
response = requests.get(url, headers=headers)
image_file = BytesIO(response.content) # Store image in memory
image = Image.open(image_file) # Open as a PIL Image object

# Display the image in your notebook or VS Code
image.show()
image
```

```python
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")



ax_image = fig.add_axes([0.1, 0.11, 0.15, 0.35])

ax_image.imshow(image)
ax_image.axis('off')
```
